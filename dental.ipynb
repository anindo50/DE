```python
from bs4 import BeautifulSoup
import requests as req
import pandas as pd

brand_name = []
product_link = []
img_url = []
product_name = []
product_info = []
description = []
item_no = []
price = []

```

```python
def process():
    url = 'https://www.pearsondental.com/catalog/product_bybrand_mfg.asp'
    resp = req.get(url)
    data = resp.text
    soup = BeautifulSoup(data,'html.parser')
    
    return soup
```

```python
def scrape_brand_names():
    soup = process()
    tr = soup.find_all('table',class_='subcat') 
    for i in tr:
        x = i.find_all('a')
        for i in x:
            c = i.text
            if c:
                brand_name.append(c)
            else:
                brand_name.append('N/A')
scrape_brand_names()
```

```python
def scrape_product_links():
    soup = process()
    tr = soup.find_all('table',class_='subcat')
    for i in tr:
        x = i.find_all('a')
        for i in x:
            b = i.get('href')
            product_link.append(b)
            link = 'https://www.pearsondental.com/catalog/' + b
            p_l = req.get(link)
            p_d = p_l.text
            p_s = BeautifulSoup(p_d,'html.parser')
            x = p_s.strong
            if x:
                b = x.text.split('-')
                product_name.append(b[1])
            else:
                product_name.append('N/A')           
    
scrape_product_links()
```

```python
def scrape_product_info():
    soup = process()
    tr = soup.find_all('table',class_='subcat')
    for i in tr:
        x = i.find_all('a')
        for i in x:
            b = i.get('href')
            link = 'https://www.pearsondental.com/catalog/' + b
            if b:
                product_link.append(link)
            else:
                product_link.append('N/A')
            p_l = req.get(link)
            p_d = p_l.text
            p_s = BeautifulSoup(p_d,'html.parser')
            img_c = p_s.find('img')
            if img_c:
                img = img_c.get('src')
                img_url.append(img)
            else:
                img_url.append('N/A')
            
            con = p_s.find('a',class_='subcat')
            if con:
                link = 'https://www.pearsondental.com' + con.get('href')
                con_l = req.get(link)
                con_d = con_l.text
                con_s = BeautifulSoup(con_d,'html.parser')
                pro_info_c = con_s.find('font',class_='link2')
                if pro_info_c:
                    pro_info = pro_info_c.text
                    product_info.append(pro_info)
                else:
                    product_info.append('N/A')
                detail = con_s.find('font',class_='link2').find_next('font').text
                if detail:
                    description.append(detail)
                else:
                    description.append('N/A')
                x = con_s.find_all('td',class_='link2')
                for i in x:
                    z = i.find('a')
                    if z:
                        q = z.get('name')
                        item_no.append(q)
                    else:
                        item_no.append('N/A')
                pr = con_s.find('font',class_='maroon').text
                if pr:
                    price.append(pr)
                else:
                    price.append('N/A')
                
                
scrape_product_info()
```

```python
def export_csv():
    #dataFrame = pd.DataFrame(data_dict)
    #dataFrame.to_csv("dental.csv", mode='w', header=True, index=False)
    bn = pd.DataFrame({'Brand Name':brand_name})
    pl = pd.DataFrame({'Product Link':product_link})
    im = pd.DataFrame({'Image Url':img_url})
    pn = pd.DataFrame({'Product Name':product_name})
    pf = pd.DataFrame({'Product_info':product_info})
    de = pd.DataFrame({'Description':description})
    it = pd.DataFrame({'Item NO':item_no})
    pr = pd.DataFrame({'Price':price})
    df = pd.concat([bn,pl,im,pn,pf,de,it,pr],axis = 1)
    df.to_csv('dent.csv',encoding = 'utf-8',header=True)
    
    return df

export_csv()
```
